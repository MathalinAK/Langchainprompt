use the gemini pro vision

```python
model = genai.GenerativeModel('gemini-1.5-flash')
```

```python
with open('flower.jpg', 'rb') as f:
    image_bytes = f.read()

# Now use image_bytes in the request
response = model.generate_content([
    {"mime_type": "image/jpeg", "data": image_bytes},  # Pass image bytes
    {"text": "Describe this image"}  # Text prompt
])
tomarkdown(response.text)
```

```python
with open('flower.jpg', 'rb') as f:
    image_bytes = f.read()
response = model.generate_content([
    {"mime_type": "image/jpeg", "data": image_bytes},
    {"text": "write a short,engaging blog post based on this picture.it should include a description "}
], stream=True)  #passing Stream parameter directly
for chunk in response:
  print(chunk.text)
```
